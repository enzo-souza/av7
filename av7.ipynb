```python
import numpy as np
import matplotlib.pyplot as plt

pontos = []
i = 0
while i < 1:
    f = np.sin(2 * np.pi * i)
    pontos.append(f)
    i += 0.01

plt.plot(pontos, marker='o', linestyle='-', color='b')
plt.title('2*pi*x')
plt.xlabel('Índice')
```

```python
i = 0
pontos = []
while i < 1:
    f = np.cos(2 * np.pi * i) * 2*np.pi
    pontos.append(f)
    i += 0.01

plt.plot(pontos, marker='o', linestyle='-', color='r')
plt.title('2*pi*cos(2*pi*x)')
plt.xlabel('Índice')
```
