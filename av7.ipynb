```python
import numpy as np

def regra_simpson(f, a, b, n):
    """
    Calcula a integral definida de uma função 'f' de 'a' até 'b'
    usando a Regra de 1/3 de Simpson com 'n' intervalos.

    Parâmetros:
    f (function): A função a ser integrada. Deve aceitar um número (float/int) e retornar um número.
    a (float): O limite inferior de integração.
    b (float): O limite superior de integração.
    n (int): O número de intervalos para a aproximação. DEVE SER UM NÚMERO PAR.
    
    Retorna:
    float: O valor aproximado da integral.
    """
    
    # Calcula a largura de cada intervalo (h)
    h = (b - a) / n

    # Soma dos termos nas extremidades
    soma = f(a) + f(b)

    # Soma dos termos com coeficiente 4 (índices ímpares)
    for i in range(1, n, 2):
        x = a + i * h
        soma += 4 * f(x)

    # Soma dos termos com coeficiente 2 (índices pares)
    for i in range(2, n, 2):
        x = a + i * h
        soma += 2 * f(x)

    # Aplica a fórmula final
    integral = (h / 3) * soma
    return integral

# --- Exemplo de Uso ---

# 1. Defina a função que você quer integrar.
#    Vamos usar f(x) = x^2, cuja integral de 0 a 3 é exatamente 9.
def minha_funcao(x):
    return np.sin(2 * np.pi * x)

# 2. Defina os limites de integração e o número de intervalos.
limite_inferior = 0
limite_superior = 1
num_intervalos = 10 # Um número par para boa precisão

# 3. Chame a função e imprima o resultado.
resultado = regra_simpson(minha_funcao, limite_inferior, limite_superior, num_intervalos)
print(f"Exemplo 1: Integral de x^2 de {limite_inferior} a {limite_superior}")
print(f"Resultado aproximado: {resultado}")
print("Resultado exato: 0.0\n")

```
